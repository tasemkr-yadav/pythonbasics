```python
# print a column

df_train['Age']
```

```python
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].mean())
```

```python
df_train['Cabin'].unique()
```

```python
#just in current window
df_train.drop(['Cabin'], axis = 1)
```

```python
# original data changes
df_train.drop(['Cabin'], axis = 1, inplace=True)
```

```python
df_train['Embarked'].unique()
```

```python
#value counts

df_train['Embarked'].value_counts()
```

```python
#filling str values now

df_train['Embarked'] = df_train['Embarked'].fillna('S')
df_train.isnull().sum()
```

```python
# WHO COULD'NT SURVIVE WITH GENDER

df_train[df_train['Survived'] == 0]['Sex'].value_counts()

# WHO COULD'NT SURVIVE WITH GENDER AND AGE < 18

df_train[(df_train['Survived'] == 0) & (df_train['Age'] < 18)]['Sex'].value_counts()

# WHO SURVIVED WHAT IS THEIR AVG TICKET PRICE

print(df_train[df_train['Survived'] == 1]['Fare'].mean())

```

```python
# Group By

print(df_train.groupby(['Survived', 'Sex'])[['Fare', 'Age']].mean())

print(df_train.groupby(['Survived'])['Sex'].count())

```

```python
#Feature Engineering

df_train['isAdult'] = 'None'
df_train.head()
```

```python
df_train['isAdult'] = np.where(df_train['Age'] >= 18, 'Adult', 'Child')
df_train.head()
```

```python
#fare bin with quantiles of fare

df_train['FareBin'] = np.where(df_train['Fare'] < 7.9, 'Low',
                               np.where((df_train['Fare'] > 7.9) & (df_train['Fare'] < 14.45), 'Medium',
                               np.where((df_train['Fare'] > 14.45), 'High', None)))
df_train.head()

```

```python
df_train.drop(['PassengerId'], axis = 1, inplace = True)
```
